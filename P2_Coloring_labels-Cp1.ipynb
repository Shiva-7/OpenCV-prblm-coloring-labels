# Write a function to highlight and fill the following Labels (Merchant ID, Transaction ID) in light blue and Its values in light red and return modified image.

```python
#importing required libraries
#cv2(Open) for solving computer vision related problems
#pytesseract for exracting text from images which used OCR technique
#PIL is an imagning library in python 
#which supports for openning manipulating and saving images in different file formats


import cv2
import pytesseract
from pytesseract import image_to_string
from PIL import Image,ImageDraw,ImageFont
```

```python
def modified_img(str):
    raw_img = cv2.imread(str)                              #reads the string and stores in raw_img
    text1 = pytesseract.image_to_string(raw_img)           #exracting text from image
    plain_img=Image.open("C:/Users/shiva/python programs/Basic projects/Thirdleap/pwbcropsm.jpg") #passing a plain image to finally show the modified image(will included a plain img while sending this file replace its path here to get the output)
    txt1 = text1[0:118]                                    #using slicing on text at the specific locations so pass the text on to the image
    txt2 = text1[118:130]
    txt2v = text1[130:141]
    txt3 = text1[142:159]
    txt4 = text1[161:176]
    txt4v = text1[177:186]
    txt5 = text1[187:]
    font_type = ImageFont.truetype('arial.ttf',12)         #fonttype Specifications
    draw = ImageDraw.Draw(plain_img)                       #the Draw funtion helps us to manipulate the image
    draw.text(xy=(100,70),text=txt1,fill=(0,0,0),font=font_type)
    draw.text(xy=(100,200),text=txt2,fill=(173,216,230),font=font_type) #Merchant ID
    draw.text(xy=(175,200),text=txt2v,fill=(255,0,0),font=font_type)    #MID value
    draw.text(xy=(100,213),text=txt3,fill=(0,0,0),font=font_type)       #terminal ID
    draw.text(xy=(100,233),text=txt4,fill=(173,216,230),font=font_type) #Transaction ID
    draw.text(xy=(192,233),text=txt4v,fill=(255,0,0),font=font_type)    #TID value
    draw.text(xy=(100,247),text=txt5,fill=(0,0,0),font=font_type)
    plain_img.show()                                       #showing the output in image format
```

```python
modified_img("C:/Users/shiva/python programs/Basic projects/Thirdleap/Assignment/2.jpg")
```

# Here is the code for same operation by using tkinter

```python
import cv2
import pytesseract
from pytesseract import image_to_string
import tkinter as tk
```

```python
def highlight_txt(str):
    load_img = cv2.imread(str)
    get_text = pytesseract.image_to_string(load_img)
    root = tk.Tk()
    text1 = get_text[0:118]
    text2 = get_text[118:130]
    text2v = get_text[130:141]
    text3 = get_text[142:159]
    text4 = get_text[161:176]
    text4v = get_text[177:186]
    text5 = get_text[187:]
    tk.Label(root,text=text1).pack()
    tk.Label(root,text=text2,bg='lightblue').pack()
    tk.Label(root,text=text2v,bg='red').pack()
    tk.Label(root,text=text3).pack()
    tk.Label(root,text=text4,bg='lightblue').pack()
    tk.Label(root,text=text4v,bg='red').pack()
    tk.Label(root,text=text5).pack()
    root.mainloop()
```

```python
highlight_txt("C:/Users/shiva/python programs/Basic projects/Thirdleap/Assignment/2.jpg")
```
